```python
import pandas as pd
```

```python
fuel_vehicles = pd.read_csv('/content/drive/MyDrive/Proyecto_Henry/PROYECTO_FINAL/EDA/Rating Fuel Consumation/Vehicles_cars_gasoline/gasoline_cars.csv')
```

```python
fuel_vehicles.drop(columns=['Electric range (km)', 'De', 'Vf','z (Wh/km)','MMS', 'Tan', 'Enedc (g/km)','W (mm)','At1 (mm)','At2 (mm)','Ernedc (g/km)', 'ech','RLFI','Cn','Ct','Cr','r'], inplace=True)
```

```python
fuel_vehicles.dropna(subset=['Fuel consumption '], inplace=True)
```

```python
name_columns = [
  'identifier',
  'country',
  'vehicle code',
  'primary manufacturer',
  'vehicle brand',
  'manufacturer',
  'type',
  'variant',
  'version',
  'make',
  'mass in running order (kg)',
  'wltp test mass',
  'emissions co2 g/km (wltp)',
  'fuel type',
  'fuel mode',
  'engine capacity (cm³)',
  'engine power (kw)',
  'innovation technology',
  'emission reduction g/km (wltp)',
  'status',
  'registration year',
  'date of registration',
  'fuel consumption'
]
old_column_names = fuel_vehicles.columns
rename_dict = dict(zip(old_column_names, name_columns))

fuel_vehicles = fuel_vehicles.rename(columns=rename_dict)
```

```python
fuel_vehicles.drop(columns=['innovation technology','emission reduction g/km (wltp)'], inplace = True)
```

```python
fuel_vehicles.dropna(subset=["wltp test mass","emissions co2 g/km (wltp)","engine capacity (cm³)","engine power (kw)"], inplace = True)
```

```python
fuel_vehicles.drop(columns=["type","variant","version"], inplace = True)
```

```python
fuel_vehicles[["wltp test mass","emissions co2 g/km (wltp)","engine capacity (cm³)","engine power (kw)"]] = fuel_vehicles[["wltp test mass","emissions co2 g/km (wltp)","engine capacity (cm³)","engine power (kw)"]].applymap(int)
```

```python
fuel_vehicles = fuel_vehicles[fuel_vehicles['fuel type'] != 'ng']
```

```python
diesel_df = fuel_vehicles[fuel_vehicles['fuel type'] == 'diesel']
petroleo_df = fuel_vehicles[fuel_vehicles['fuel type'] == 'petrol']
e85_df = fuel_vehicles[fuel_vehicles['fuel type'] == 'e85']

diesel_df_len = len(diesel_df)
petroleo_df_len = len(petroleo_df)
e85_df_len = len(e85_df)

cantidad_deseada_cada_tipo_gasolina = 31115

total_filas_diesel = diesel_df_len - (diesel_df_len - cantidad_deseada_cada_tipo_gasolina)
total_filas_petroleo = petroleo_df_len - (petroleo_df_len - cantidad_deseada_cada_tipo_gasolina)
total_filas_e85 = e85_df_len -  (e85_df_len - cantidad_deseada_cada_tipo_gasolina)

diesel_df = diesel_df.head(total_filas_diesel)
petroleo_df = petroleo_df.head(total_filas_petroleo)
e85_df = e85_df.head(total_filas_e85)

dataset_balanceado_fuel_vehicle = pd.concat([diesel_df, petroleo_df, e85_df])
```
