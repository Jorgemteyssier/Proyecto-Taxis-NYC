```python
import pandas as pd
```

```python
df_hybrid_electric_vehicles = pd.read_csv('/content/drive/MyDrive/Proyecto_Henry/PROYECTO_FINAL/EDA/Rating Fuel Consumation/Plug-_in_hybrid_electric_vehicles_Datasets/hybrid_vehicle.csv')
```

```python
df_hybrid_electric_vehicles.drop(columns=['De', 'Vf','MMS', 'Tan', 'Enedc (g/km)','W (mm)','At1 (mm)','At2 (mm)','Ernedc (g/km)', 'ech','RLFI','Cn','Ct','Cr','r'], inplace=True)
```

```python
name_columns = [
  'identifier',
  'country',
  'vehicle code',
  'primary manufacturer',
  'vehicle brand',
  'manufacturer',
  'type',
  'variant',
  'version',
  'make',
  'mass in running order (kg)',
  'wltp test mass',
  'emissions co2 g/km (wltp)',
  'fuel type',
  'fuel mode',
  'engine capacity (cmÂ³)',
  'engine power (kw)',
  'energy_consumption_wh_per_km',
  'innovation technology',
  'emission reduction g/km (wltp)',
  'status',
  'registration year',
  'date of registration',
  'fuel consumption',
  'electric range (km)'
]

old_column_names = df_hybrid_electric_vehicles.columns
rename_dict = dict(zip(old_column_names, name_columns))

df_hybrid_electric_vehicles = df_hybrid_electric_vehicles.rename(columns=rename_dict)
```

```python
df_hybrid_electric_vehicles.drop(columns=['innovation technology','emission reduction g/km (wltp)'], inplace = True)
```

```python
df_hybrid_electric_vehicles.drop(columns=["type","variant","version"], inplace = True)
```

```python
df_hybrid_electric_vehicles["fuel type"] = df_hybrid_electric_vehicles["fuel type"].str.lower()
```

```python
fuel_counts = df_hybrid_electric_vehicles["fuel type"].value_counts()

min_count = fuel_counts.min()

df_petrol = df_hybrid_electric_vehicles[df_hybrid_electric_vehicles["fuel type"] == 'petrol/electric'].sample(min_count, random_state=42)
df_diesel = df_hybrid_electric_vehicles[df_hybrid_electric_vehicles["fuel type"] == 'diesel/electric'].sample(min_count, random_state=42)


df_balanced_vehicles_hybrid = pd.concat([df_petrol, df_diesel])
```
