```python
df_electric_Cars = pd.read_csv('/content/drive/MyDrive/Proyecto_Henry/PROYECTO_FINAL/EDA/Rating Fuel Consumation/Batery_Vehicle_Consumation_Datasets/electric_cars.csv')
```

```python
name_columns = ['identifier', 'country', 'vehicle_code', 'primary manufacturer','vehicle brand','manufacturer', 'model', 'model_name', 'fuel_type', 'fuel_mode', 'power_kw', 'energy_consumption_wh_per_km', 'status', 'year', 'registration_date', 'electric_range_km']

old_column_names = df_electric_Cars.columns
rename_dict = dict(zip(old_column_names, name_columns))

df_electric_Cars = df_electric_Cars.rename(columns=rename_dict)
```

```python
df_electric_Cars2 = df_electric_Cars.copy()
```
