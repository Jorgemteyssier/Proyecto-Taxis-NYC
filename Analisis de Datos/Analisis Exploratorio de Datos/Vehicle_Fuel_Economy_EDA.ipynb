```python
#Importamos las librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Se carga el dataset
data = pd.read_csv(r"Vehicle Fuel Economy Data.csv")
data
```

```python
#Revisamos las columnas y su tipo de dato
data.info()
```

```python
#Estadísticas descriptivas para columnas numéricas
data.describe()
```

```python
#Verificamos los valores nulos
data.isnull().sum()
```

```python
#Observamos la cantidad de valores únicos en las columnas "Manufacturer", "Model", "fuelType1"
print(f"Cantidad de fabricantes únicos: {data['Manufacturer'].nunique()}") 
print(f"Cantidad de modelos únicos: {data['Model'].nunique()}")
print(f"Tipos de combustibles: {data['fuelType1'].unique()}")   
```

```python
#Visualizamos valores nulos
plt.figure(figsize=(10,6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de valores nulos')
plt.show()
```

Las columnas con muchos valores nulos pueden necesitar mas atención al detalle, ya sea mediante reemplazo de datos o eliminación de esas columnas.
Es importante tener en cuenta estos valores nulos en cualquier análisis que se quiera realizar, ya que pueden afectar los resultados y conclusiones.

```python
#Histograma para la variable 'city08' (rendimiento en ciudad)
plt.figure(figsize=(10,6))
sns.histplot(data['city08'], kde=True, bins=30)
plt.title('Distribución del rendimiento en ciudad')
plt.show()

```

Grafico Distribución de rendimiento en ciudad: La mayoría de los vehículos tienen un rendimiento en ciudad concentrado entre 10 y 30 mpg, con un pico alrededor de 20 mpg.
La cola hacia la derecha indica que hay algunos vehículos que tienen un rendimiento mucho mayor, pero son mucho menos frecuentes.

```python
#Histograma para la variable 'co2' (emisiones de CO2)
plt.figure(figsize=(10,6))
sns.histplot(data['co2'], kde=True, bins=30)
plt.title('Distribución de emisiones de CO2')
plt.show()
```

Grafico Distribución de Emisiones CO2 muestra una distribución altamente sesgada hacia la izquierda, indicando que la mayoría de los vehículos tienen emisiones de CO2 muy bajas, probablemente eléctricos o híbridos. Hay un pequeño grupo con emisiones más altas, alrededor de 300-400 g/km, representando vehículos menos eficientes.

```python
#Boxplot para 'co2' según el tipo de combustible
plt.figure(figsize=(10,6))
sns.boxplot(x='fuelType1', y='co2', data=data)
plt.title('Emisiones de CO2 por tipo de combustible')
plt.show()

```

Los vehículos que utilizan gasolina premium tienen una mayor variabilidad en las emisiones, con valores que llegan hasta casi 900 g/km. Gasolina regular muestra una gran cantidad de vehículos con bajas emisiones y muchos valores atípicos, reflejando una diversidad en la eficiencia de los vehículos. Diesel tiene emisiones medianas pero más consistentes, mientras que los vehículos eléctricos y de gas natural prácticamente no emiten CO2. gasolina de grado medio muestra una distribución moderada con pocos valores extremos.

```python
#Boxplot para 'city08' según el tipo de combustible
plt.figure(figsize=(10,6))
sns.boxplot(x='fuelType1', y='city08', data=data)
plt.title('Rendimiento en ciudad por tipo de combustible')
plt.show()
```

Este gráfico muestra el rendimiento de los vehículos en ciudad según el tipo de combustible, medido en millas por galón. Los vehículos eléctricos tienen el mejor rendimiento. Los vehículos que usan gasolina regular, gasolina premium, y gasolina de grado medio tienen rendimientos similares y más bajos, entre 15 y 30 millas por galón, con varios valores atípicos. Diesel y gas natural muestran un rendimiento más estable y algo mejor, aunque inferior al de los vehículos eléctricos.

```python
#Filtramos las columnas que consideramos utiles para nuestro proyecto 
columnas_utilizar= ["Year", "Manufacturer", "Model", "barrels08", "city08", "city08U", "co2", "co2TailpipeGpm", "comb08", 
                  "engId", "fuelCost08", "fuelType1", "ghgScore", "highway08", "id", "UCity"]
data_filtrado = data[columnas_utilizar]
```

```python
data_filtrado
```

```python
#Revisamos nuevamente los valores nulos de este dataset filtrado
data_filtrado.isnull().sum()
```

```python
#Rellenar columnas numéricas con 0
columnas_numericas = ["barrels08", "city08", "city08U", "co2", "co2TailpipeGpm", "comb08", "engId", "fuelCost08", "ghgScore", "highway08", "id", "UCity"]
data_filtrado[columnas_numericas] = data_filtrado[columnas_numericas].fillna(0)
```

```python
#Reemplazamos los NaN de las columnas de texto por "Sin dato"
columnas_categoricas = ["Model", "fuelType1"]
data_filtrado[columnas_categoricas] = data_filtrado[columnas_categoricas].fillna("Sin dato")
```

```python
#Revisión de los datos después de reemplazar los valores nulos
data_filtrado.isnull().sum()
```

```python
data_filtrado
```

```python
#Transformamos todos los caracteres de la columna "Model", "Manufacturer" y "fuelType1" a minúsculasa minúsculas
columnas = ["Model", "Manufacturer", "fuelType1"]
data_filtrado[columnas] = data_filtrado[columnas].apply(lambda x: x.str.lower())

#Verificamos los cambios
data_filtrado
```

```python
#Revisamos los valores unicos de la columna ID
print(f"Cantidad de id únicos: {data_filtrado['id'].nunique()}") 
```

```python
#Verificamos los valores duplicados de la columna ID
data_filtrado['id'].duplicated().sum()
```

```python
#Eliminamos los valores duplicados en la columna ID
data_filtrado = data_filtrado.drop_duplicates(subset=['id'])
```
